Consideremos una distribución gaussiana de dos variables $x$, $y$ dada por

$$
\rho(x,y) = \frac{\sqrt{c_{xx}c_{yy} - c_{xy}^2}}{2\pi}\exp\left[-\frac{1}{2}c_{xx} (x - \mu_x)^2 - c_{xy} (x - \mu_x)(y - \mu_y) - \frac{1}{2} c_{yy} (y - \mu_y)^2\right]
$$

La probabilidad de obtener un valor dado de $x$ independiente del valor de $y$ está dada por

$$
\rho(x) = \int_{-\infty}^\infty dy\,\rho(x,y)\,,
$$

porque sumamos sobre todos los posibles valores de $y$. A esto se lo llama "marginalizar sobre y".

Considere el caso $c_{xx} = 2$, $c_{yy} = 1$, $c_{xy} = 1.4$, $\mu_x = 3$, $\mu_y = 4$.

* Grafique las distribuciones de probabilidad $\rho(x)$ y $\rho(y)$.

* Grafique la distribución en 3 dimensiones (el eje $z$ es el valor de $\rho$).

* Suponga que esta probabilidad representa el resultado de un experimento que a partir de ciertos datos asigna una probabilidad a las cantidades $x$ y $y$. Discuta qué se puede concluir sobre el punto $x = 0$, $y = 0$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Reemplazando los valores para $\rho(x,y)$ e integrando para x e y se obtiene:

$$
\rho(x) =\frac{\exp[\frac{-(x-3)^2}{50}]}{5\sqrt{2\pi}}
$$


$$
\rho(y) =\frac{\exp[\frac{-(y-4)^{2}}{100}]}{\sqrt{50}\sqrt{2\pi}}
$$


Y por la formula se puede obtener $\sigma_y = \sqrt{50}$ y $\sigma_x = 5$

```python
def gauss(x, mu, sigma):                 #definimos la gaussiana p(x)
    return np.exp(-(x - mu)**2/(2*sigma**2))/(np.sqrt(2*np.pi)*sigma)
```

```python
x = np.linspace(-30, 30, 100)                  #escogemos valores arbitrarios para las variables, en este caso desde -30 a 30 para poder apreciar bien la curva gaussiana
y = np.linspace(-30, 30, 100)

plt.plot(x, gauss(x, 3, 5), label="p(x)")
plt.plot(y, gauss(y, 4, np.sqrt(50)), label="p(y)")              #graficamos la distribucion de probabilidad en funcion de las variables

plt.xlabel("variable")
plt.ylabel("Distribucion de probabilidad")
plt.legend()
plt.show()
```

```python
from mpl_toolkits.mplot3d import axes3d  #importamos el módulo que nos permite graficar en 3d

z = (0.1/np.pi)*np.exp(-(x - 3)**2 - 1.4*(x- 3)*(y - 4) - (1/2)*(y - 4)**2)  #definimos el eje z como la funcion p(x,y)


fig = plt.figure()                                #graficamos
ax1 = fig.add_subplot(111,projection='3d')          

plt.xlabel("x")
plt.ylabel("y")
plt.title("p(x,y)")
ax1.plot(x, y, z)
plt.show()
```

El punto 0,0 es donde la distribución de probabilidad empieza a aumentar
