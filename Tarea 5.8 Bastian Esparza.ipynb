```python
import numpy as np
import matplotlib.pyplot as plt
```

Para este problema usaremos la formula de poisson\\
$$
P = ((e^{-\lambda})\lambda^{x})/x!
$$

```python
def probabilidad(n, l):                                    #definimos la funcion para n estrellas fugaces
    return(np.exp(-l)*l**n/np.math.factorial(n))

x = int(input("x="))    #se le pedira al usuario un numero para graficar
n = np.arange(0, x+1, 1)                                      #definimos n como un arreglo de 0 a n estrellas
p = np.array([poisson(n, 15.7/2) for n in range(x+1)])  

plt.plot(n, p)        #Graficamos los datos.
plt.title('Probabilidad de observar n estrellas en media hora')
plt.xlabel('n estrellas')
plt.ylabel('probabilidad')
plt.show()
```
